# Similar Breakfast Cereal

### Part 2: Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data with normalized k-means clusters
cereal = pd.read_csv('norm.csv')
```

```python
cereal
```

```python
# duplicate kmeans_norm_labels column
cereal['cluster']=cereal['kmeans_norm_labels']
```

```python
cereal
```

```python
# Reassign cluster labels
cereal['cluster']=cereal['cluster'].replace({0:'D', 1:'A', 2:'E', 3:'B', 4:'C', 5:'F'})
```

```python
cereal
```

```python
# sort clusters in alpha order
cereal = cereal.sort_values('cluster')
```

```python
cereal
```

#### Histograms of features for each cluster

```python
fig, axes = plt.subplots(1, 6, figsize=(10,2.5), dpi=100, sharex=True, sharey=True)

colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink', 'tab:olive', 'tab:orange']

for i, (ax, cluster) in enumerate(zip(axes.flatten(), cereal.cluster.unique())):
    x = cereal.loc[cereal.cluster==cluster, 'calories']
    
    ax.hist(x, alpha=0.8, bins=5, density=True, stacked = True, label = str(cluster), color=colors[i])
    
    ax.set_title(cluster)

plt.suptitle('Histogram of Calories per Cluster', y = 1.05, size=16)

plt.tight_layout();
```
